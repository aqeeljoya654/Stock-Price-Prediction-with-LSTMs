```python
# ==========================================
# 1. Imports
# ==========================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from sklearn.metrics import mean_squared_error
import math

# ==========================================
# 2. Download Stock Data
# ==========================================
ticker = "AAPL"   # 👈 Change this to "TSLA", "MSFT", "AMZN", etc.
start_date = "2015-01-01"
end_date = "2023-12-31"

df = yf.download(ticker, start=start_date, end=end_date)
print("Dataset shape:", df.shape)
df.head()

# ==========================================
# 3. Exploratory Data Analysis (EDA)
# ==========================================
plt.figure(figsize=(12,6))
plt.plot(df["Close"], label="Closing Price")
plt.title(f"{ticker} Stock Closing Price")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.show()

# Moving Averages
df["MA50"] = df["Close"].rolling(50).mean()
df["MA200"] = df["Close"].rolling(200).mean()

plt.figure(figsize=(12,6))
plt.plot(df["Close"], label="Closing Price")
plt.plot(df["MA50"], label="50-day MA", alpha=0.8)
plt.plot(df["MA200"], label="200-day MA", alpha=0.8)
plt.title(f"{ticker} Closing Price with Moving Averages")
plt.legend()
plt.show()

# ==========================================
# 4. Data Preprocessing
# ==========================================
data = df[["Close"]].copy()
dataset = data.values

training_data_len = int(len(dataset) * 0.8)

# Scale data between 0 and 1
scaler = MinMaxScaler(feature_range=(0,1))
scaled_data = scaler.fit_transform(dataset)

# Create training dataset
train_data = scaled_data[0:training_data_len, :]
x_train, y_train = [], []
for i in range(60, len(train_data)):
    x_train.append(train_data[i-60:i, 0])
    y_train.append(train_data[i, 0])

x_train, y_train = np.array(x_train), np.array(y_train)
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))

# ==========================================
# 5. Build LSTM Model
# ==========================================
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(x_train.shape[1], 1)))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=False))
model.add(Dropout(0.2))
model.add(Dense(units=25))
model.add(Dense(units=1))

model.compile(optimizer="adam", loss="mean_squared_error")

# Train model
history = model.fit(x_train, y_train, batch_size=32, epochs=5)

# ==========================================
# 6. Testing Model
# ==========================================
test_data = scaled_data[training_data_len - 60:, :]
x_test, y_test = [], dataset[training_data_len:, :]
for i in range(60, len(test_data)):
    x_test.append(test_data[i-60:i, 0])

x_test = np.array(x_test)
x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))

predictions = model.predict(x_test)
predictions = scaler.inverse_transform(predictions)

rmse = math.sqrt(mean_squared_error(y_test, predictions))
print("RMSE:", rmse)

# ==========================================
# 7. Visualize Predictions
# ==========================================
train = data[:training_data_len]
valid = data[training_data_len:]
valid["Predictions"] = predictions

plt.figure(figsize=(12,6))
plt.plot(train["Close"], label="Train Data")
plt.plot(valid["Close"], label="Actual Price")
plt.plot(valid["Predictions"], label="Predicted Price")
plt.title(f"{ticker} Stock Price Prediction (LSTM)")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.show()

# ==========================================
# 8. Forecast Next 30 Days
# ==========================================
last_60_days = scaled_data[-60:]
forecast_input = last_60_days.reshape(1, -1)
forecast_input = forecast_input.tolist()[0]

future_predictions = []
for i in range(30):  # forecast next 30 days
    X = np.array(forecast_input[-60:])
    X = X.reshape(1, 60, 1)
    pred = model.predict(X)[0][0]
    forecast_input.append(pred)
    future_predictions.append(pred)

future_predictions = scaler.inverse_transform(np.array(future_predictions).reshape(-1,1))

plt.figure(figsize=(12,6))
plt.plot(range(len(dataset)), dataset, label="Historical Data")
plt.plot(range(len(dataset), len(dataset)+30), future_predictions, label="Future Forecast", color="red")
plt.title(f"{ticker} Stock Price Forecast (Next 30 Days)")
plt.legend()
plt.show()
```
